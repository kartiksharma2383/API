```python
import requests
```

```python
r = requests.get('https://xkcd.com/353/')
print(dir(r))  # For content of file
```

```python
print(help(r)) # For detail explanation
```

```python
print(r.text)  # For HTML content
```

```python
print(r.status_code) # 200 output means request is successfull.
```

```python
# True output means status code successfully runs, if it doesn't runs it shows false
print(r.ok)
```

```python
print(r.headers) # It shows all headers of the content
```

```python
# payload = {'page': 2, 'count': 25}: This line creates a Python dictionary named payload. This dictionary holds key-value pairs that will be sent as query parameters in the URL for GET request.

# params = payload : In this the requests library takes the key-value pairs from your payload dictionary and appends them to the URL as query string parameters for filtering the data.
payload = {'page': 2, 'count': 25}
r = requests.get('https://httpbin.org/get', params = payload)
print(r.text)
```

```python
print(r.url)
```

```python
# payload = {'username': 'corey', 'password': 'testing'}: This line creates a Python dictionary named payload. This dictionary holds key-value pairs.(username, password) that will be sent as query parameters in the URL for POST request.

# data = payload: In this the requests library takes the key-value pairs from your payload dictionary and encodes them (usually as application/x-www-form-urlencoded by default) into the body of the HTTP request.

payload = {'username': 'corey', 'password': 'testing'}
r = requests.post('https://httpbin.org/post', data = payload)
print(r.text)
```

```python
payload = {'username': 'corey', 'password': 'testing'}
r = requests.post('https://httpbin.org/post', data = payload)
r_dict = r.json() # Convert HTTP content in Python dictionary (r_dict).
print(r_dict['form'])
```

```python
# auth is used to send credentials, and the tuple holds them as a (username, password) pair
r = requests.get(
    'https://httpbin.org/basic-auth/corey/testing',
    auth = ('corey', 'testing')
)
print(r.text)
```

```python
# /delay/1: server waits 1 second, then responds
# timeout = 3: Wait max 3 seconds for response, otherwise stop
try:
  r = requests.get('https://httpbin.org/delay/1', timeout = 3)
  print(r)
except requests.exceptions.Timeout:
  print("Timeout error")
except requests.exceptions.HTTPError:
  print("HTTP error")
except Exception as e:
  print("Other error:", e)
```
